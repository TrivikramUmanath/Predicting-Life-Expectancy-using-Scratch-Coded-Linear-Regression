```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
life=pd.read_csv("C:\\Users\\Trivikram\\ML_Projects\\Data\\Life Expectancy Data.csv")
life.columns=['Country', 'Year', 'Status', 'Life Expectancy', 'Adult Mortality',
       'Infant Deaths', 'Alcohol', 'Percentage Expenditure', 'Hepatitis B',
       'Measles', ' BMI', 'Under-Five Deaths', 'Polio', 'Total Expenditure',
       'Diphtheria','HIV/AIDS', 'GDP', 'Population',
       'Thinness  1-19 years', 'Thinness 5-9 years',
       'Income composition of resources', 'Schooling']
life
```

```python
def LinearBatchDescent(X,y,Constants,learningRate,limit):
    out=y.values
    previousMSE=np.inf
    length=len(X.columns)
    delta=np.zeros(length+1)
    MSE=0
    size=X.shape[0]
    flag=0
    while previousMSE-MSE>=limit:
        if flag==0:
            pass
        else:
            previousMSE=MSE
        slopes=Constants[:length]
        c=Constants[length]
        Ypred=X.apply(lambda x: np.sum(np.multiply(x,slopes))+c,axis=1)
        diff=Ypred-y
        MSE=(diff**2).sum()/size
        delta[:length]=X.apply(lambda x:np.sum(2*x*(diff))/size,axis=0).values.tolist()
        delta[length]=2*(Ypred-y).sum()/size
        prevConstants=Constants
        Constants=Constants-delta*learningRate
        flag=1
    return previousMSE,prevConstants

def LinearPrediction(X,y,Constants):
    out=y.values
    length=len(X.columns)
    slopes=Constants[:length]
    c=Constants[length]
    size=X.shape[0]
    Ypred=X.apply(lambda x: np.sum(np.multiply(x,slopes))+c,axis=1)
    MSE=(((Ypred-out)**2).sum())/size
    return MSE
```

```python
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
imputedData=imp_median.fit_transform(life[['Life Expectancy','Alcohol','Hepatitis B','Polio','Diphtheria']])
life[['Life Expectancy','Alcohol','Hepatitis B','Polio','Diphtheria']]=imputedData
print(life.isna().sum())
```

```python
scaler = StandardScaler()
X=life[['Alcohol', 'Hepatitis B','Measles','Polio','Diphtheria', 'HIV/AIDS']]
X=pd.DataFrame(data=scaler.fit_transform(X),columns=X.columns)
y=life['Life Expectancy']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
length=len(X.columns)+1
Constants=np.random.rand(length)
learningRate=float(input("Enter the learning Rate you intend to use "))
limit=float(input("Enter the threshold difference between the previousMSE and Current MSE "))
trainMSE,Constants=LinearBatchDescent(X_train,y_train,Constants,learningRate,limit)
testMSE=LinearPrediction(X_test,y_test,Constants)
print("The trainiMSE is "+str(trainMSE)+" and the testMSE is "+str(testMSE))
print("The constants to be used are "+ str(Constants))
```

```python
linearRegressor=LinearRegression()
linearRegressor.fit(X_train,y_train)
Predictions=np.asarray(linearRegressor.predict(X_test))
ActualValues=y_test.values
MSE=((Predictions-ActualValues)**2).sum()/y_test.shape[0]
print("The MSE obtained after Linear Regression is "+str(MSE))
print("The coefficents are "+str(linearRegressor.coef_))
print("The intercept is "+str(linearRegressor.intercept_))
```
