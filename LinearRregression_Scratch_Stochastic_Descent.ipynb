```python
m=float(input("Enter the value of m "))
c=float(input("Enter the value of c "))
learningRates=[0.01]
iterations=int(input("Enter the number of iterations "))
size=df.shape[0]
inp=df['X']
out=df['Y']
MSEs=[]
for learningRate in learningRates:
    for l in range(iterations):
        print("Iteration "+str(l+1))
        print(learningRate)
        Ypred=[m*i+c for i in inp]
        MSE=sum([(j-k)**2 for (j,k) in zip(Ypred,out)])/size
        deltaM=sum([2*i*(j-k) for (i,j,k) in zip(inp,Ypred,out)])/size
        deltaC=sum([2*(j-k) for (j,k) in zip(Ypred,out)])/size
        m=m-deltaM*learningRate
        c=c-deltaC*learningRate
        print("MSE is "+str(MSE)+", M is "+str(m)+" and c is "+str(c))
        if(l==0):
            previousMSE=MSE
        if(MSE<=previousMSE):
            previousMSE=MSE
        else:
            print("Overshoot")
            break
    MSEs.append(previousMSE)
```
